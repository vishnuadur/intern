1.Python program to display all the header tags from wikipedia.org.

# 2)   Write a python program to display IMDB’s Top rated 100 movies’ 


```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
soup = BeautifulSoup(page.content)
title=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    title.append(i.text.replace("\n",""))
title
rating=[]
for i in soup.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace("\n",""))
rating
import pandas as pd
df=pd.DataFrame({'Title and realease year':title,'Rating':rating})
df.set_index('Title and realease year')
```

9)   Write a python program to scrape mentioned details from dineout.co.in :


```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content)
name=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text.replace("\n",""))
name
cuisine=[]
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text.replace("\n","").split()[-2:])
cuisine
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text.replace("\n",""))
location

rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text.replace("\n",""))
rating
imgs=[]
for i in soup.find_all('img',class_="no-img"):
    imgs.append(i['data-src'])
imgs
import pandas as pd
df=pd.DataFrame({'Resturent name':name,'Cuisine':cuisine,'Location':location,'Rating':rating,'Image_url':imgs})
df.set_index('Resturent name')
```
